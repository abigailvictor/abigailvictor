*Modelling*

using logistic regression classifier model to classify our data

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
train.drop(columns='target')
```

```python
x_train=train.drop(columns='target')
y_train=train.target
```

```python
#logistic regression
clf=LogisticRegressionCV(max_iter=1000)  #ouclass_weight=r classifier
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_train,y_train)
```
